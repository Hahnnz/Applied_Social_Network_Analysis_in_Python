We will talk about how to find important nodes from a network.

### Node Importance

Based on the structure of the network, which are the 5 most important node in the Karate. Club friendship network?<br>
There is many different ways of thinking about **"importance"**. <br>

- Examples. (informer!)
1. Degree : Number of friends. -> 5 most important nodes are : 34, 1 33, 3, 2
2. Average Proximity to other nodes -> 5 most important nodes are : 1, 3, 34, 32, 9
3. Fraction of shortest paths that pass through node -> 5 most important nodes are : 1, 34, 33, 3, 32

### Network Centrality
Centrality measures identify the most important nodes in a network.
* Influential nodes, which has a strong infleunce to spread out, in a network (ex. Disease)
* Nodes taht

```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline

# This toy graph is so famous that "networkx" has it.
G = nx.karate_club_graph()

# Set start node number 1
G = nx.convert_node_labels_to_integers(G, first_label=1)

pos = nx.spring_layout(G, seed=201703) 
nx.draw_networkx(G,pos, node_color='white', edgecolors='black')

# get all possible node degrees from given graph, and sorted by number of degrees.
all_node_degree = dict(nx.degree(G))
ranked_by_degree = sorted(range(1,len(all_node_degree)+1), key=lambda k: all_node_degree[k], reverse=True)[:5]

# Red Nodes stand for 5 most important nodes ranked by degree
for n in ranked_by_degree :
    nx.draw_networkx_nodes([n], pos, node_color='red', alpha=0.4)
    
# get all possible node proximity from given graph.
Average_proximity = nx.closeness_centrality(G)
ranked_by_proximity = sorted(range(1,len(Average_proximity)+1), key=lambda k: Average_proximity[k], reverse=True)[:5]
    
# Blue Nodes stand for 5 most important nodes ranked by proximity
for n in ranked_by_proximity :
    nx.draw_networkx_nodes([n], pos, node_color='blue', alpha=0.4)

print(f'5 most important nodes ranked by degree : {ranked_by_degree}')
print(f'5 most important nodes ranked by proximity : {ranked_by_proximity}')
plt.show()
```
